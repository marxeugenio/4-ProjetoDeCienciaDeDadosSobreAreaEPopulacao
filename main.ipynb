```python
import pandas as pd
cidades = pd.read_csv(r"C:\Users\marxe\OneDrive\Área de Trabalho\170 Projetos\5. Projeto de Ciência de Dados sobre Área e População\california_cities.csv")
print(cidades.head())
```

```python
# extraindo os dados que nos interessa 

latitude, longitude = cidades["latd"], cidades["longd"]
populacao, area = cidades["population_total"], cidades["area_total_km2"]
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn
seaborn.set()
plt.scatter(longitude, latitude, label=None, c=np.log10(populacao),
            cmap="viridis", s=area, linewidth=0, alpha=0.5)
plt.axis(aspect="equal")
plt.xlabel("longitude")
plt.ylabel("latitude")
plt.colorbar(label="log$_{10}$(populacao)")
plt.clim(3, 7)

# agora vamos criar uma legenda, vamos plotar listas vazias com o tamanho e rótulo desejados

for area in [100, 300, 500]:
    plt.scatter([],[], c="k", alpha=0.3, s=area, label=str(area) + "km$^2$")
plt.legend(scatterpoints=1, frameon=False, labelspacing=1, title="Areas da Cidade")
plt.title("Area da População da California")
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn
seaborn.set()

# Supondo que você tenha definido 'longitude', 'latitude', 'populacao' e 'area' em algum lugar do seu código

plt.scatter(longitude, latitude, label=None, c=np.log10(populacao),
            cmap="viridis", s=area, linewidth=0, alpha=0.5)

# Correção para o ajuste de aspecto
plt.gca().set_aspect('equal', adjustable='box')

plt.xlabel("longitude")
plt.ylabel("latitude")
plt.colorbar(label="log$_{10}$(população)")
plt.clim(3, 7)

# Crie um gráfico de dispersão vazio para a legenda
for area_val in [100, 300, 500]:
    plt.scatter([], [], c="k", alpha=0.3, s=area_val, label=str(area_val) + "km$^2$")

# Use 'handles' para especificar a legenda
plt.legend(scatterpoints=1, frameon=False, labelspacing=1, title="Áreas da Cidade")

plt.title("Área da População da Califórnia")
plt.show()
```
